```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Salary_Data.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
def cost_function(x,y,w,b):
  n=x.shape[0]
  cost=0
  for i in range(n):
    cost+=(((w*x[i]-b)-y[i])**2)

  cost=cost/(2*n)

  return cost
```

```python
plt.scatter(x_train,y_train,color='red')
plt.show()
```

```python
def gradient_function(x,y,w,b):
  n=x.shape[0]
  grad_w=0
  grad_b=0

  for i in range(n):
    temp=(w*x[i]+b-y[i])
    grad_w+=temp*x[i]
    grad_b+=temp

  grad_w=grad_w/n
  grad_b=grad_b/n

  return grad_w,grad_b
```

```python
def gradient_descent(x,y,w,b,alpha):

  for i in range(10000):
    grad_w,grad_b=gradient_function(x,y,w,b)
    w=w-alpha*grad_w
    b=b-alpha*grad_b

  return w,b
```

```python

w,b=gradient_descent(x_train,y_train,0,0,1.0e-2)
print(w,b)

plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,w*x_train+b)
plt.show()


y_pred=w*x_test + b
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
w = model.coef_[0]  # Slope of the regression line (w)
b = model.intercept_  # Intercept of the regression line (b)

print(f"Slope (w): {w:.2f}")
print(f"Intercept (b): {b:.2f}")
```
