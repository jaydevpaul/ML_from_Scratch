```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('/content/Position_Salaries.csv')
```

```python
x=data.iloc[:,1:-1]
y=data.iloc[:,-1]
```

```python
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=2)
x_poly=poly_reg.fit_transform(x)
x_poly
```

```python
lin_reg2=LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
plt.scatter(x,y,c="Red")
plt.plot(x,lin_reg.predict(x),c="Blue")
# plt.plot(x_poly,lin_reg2.predict(poly_reg.fit_transform(x)),c="green")
plt.show()
```

```python
plt.scatter(x,y,c="Red")
plt.plot(x,lin_reg.predict(x),c="Blue")
plt.plot(x,lin_reg2.predict(x_poly),c="green")
plt.show()
```

```python
print(lin_reg.coef_)
print(lin_reg2.coef_)
```

```python
lin_reg3=LinearRegression();
poly=PolynomialFeatures(degree=4)
x_poly=poly.fit_transform(x)
lin_reg3.fit(x_poly,y)
```

```python
plt.scatter(x,y,c="Red")
plt.plot(x,lin_reg.predict(x),c="Blue")
plt.plot(x,lin_reg3.predict(x_poly),c="green")
plt.show()
```
